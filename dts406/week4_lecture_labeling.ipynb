```python
import torch
import torchvision
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
import numpy as np
from transformers import AutoTokenizer, AutoModelForTokenClassification
```

```python
if torch.backends.mps.is_available():
    device = torch.device("mps")
    print("Apple mps is available😀")
elif torch.cuda.is_available():
    device = torch.device("cuda")
    print ("Using cuda")
else:
    device = torch.device("cpu")
    print("Using cpu")

print (f"Current device={device}")
```

```python
model_name = "dbmdz/bert-base-cased-finetuned-conll03-english"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForTokenClassification.from_pretrained(model_name)
```

```python
id2label = model.config.id2label
```

```python
sentence = "The quick brown fox, jumped over the lazy hen"
```

```python
inputs = tokenizer(
    sentence,
    return_tensors = "pt",
    truncation = True,
    padding = True
)

print("tokenizer initialized")
```

```python
input_ids = inputs["input_ids"]
attention_mask = inputs["input_ids"]
```

```python
print("running pos labelling")
with torch.no_grad():
    outputs = model(input_ids=input_ids, attention_mask=attention_mask)

predictions = torch.argmax(outputs.logits, dim=2)
tokens = tokenizer.convert_ids_to_tokens(input_ids[0])
```

```python
results = []
for token, pred_id in zip(tokens, predictions[0]):
    if token not in ("[CLS]", "[SEP]"):
        pos_label = id2label[pred_id.item()]
        results.append ((token, pos_label))

print(f"input sentence:\n{sentence}\n")
print("Labeling resutls")

for token, pos in results:
    print(f"{token:<10} = {pos}")
```
