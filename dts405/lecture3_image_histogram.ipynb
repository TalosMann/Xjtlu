```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread("images/2.png", cv2.IMREAD_COLOR)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_grey = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)
```

```python
gray_hist = cv2.calcHist([image_grey], [0],None, [256], [0,256])
```

```python
plt.figure(figsize=(10, 10))
plt.subplot(2, 2, 1)
plt.imshow(image_grey, cmap="gray")
plt.axis("off")
plt.show

plt.subplot(2, 2, 2)
plt.plot(gray_hist, color="black")
plt.xlabel("Intensity")
plt.ylabel("Number of Pixels")
plt.xlim([0, 256])

plt.subplot(2, 2, 3)
plt.imshow(image_rgb)
plt.axis("off")

plt.subplot(2, 2, 4)
colors = ("red", "green", "blue")
channel_ids = [0, 1, 2]

for channel_id, color in zip(channel_ids, colors):
    color_hist = cv2.calcHist([image_rgb], [channel_id], None, [256], [0, 256])
    plt.plot(color_hist, color=color, label=f"{color.capitalize()}")
plt.xlabel("Intensity")
plt.ylabel("Number of Pixels")
plt.xlim([0, 256])
plt.legend()
plt.tight_layout()
plt.show
```
