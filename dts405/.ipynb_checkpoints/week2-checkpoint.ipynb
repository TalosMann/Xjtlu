```python
# Install dependencies (run once if not installed)
!pip install opencv-python numpy matplotlib --quiet

# Import core libraries
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
import os

# Set Matplotlib to display plots inline
%matplotlib inline

# --------------------------
# CONFIGURE THIS FOR YOUR FILE!
# --------------------------
IMAGE_PATH = "images/1.png"  # Replace with your image path (e.g., "images/lecture2_img.png")

# --------------------------
# Load image from local folder
# --------------------------
def load_local_image(path):
    # Check if file exists
    if not os.path.exists(path):
        raise FileNotFoundError(
            f"❌ Image not found at: {os.path.abspath(path)}\n"
            f"Example valid paths: 'images/photo.png' or 'images/lecture_img.jpg'"
        )
    
    # Load as grayscale (best for quantization)
    img = cv.imread(path, cv.IMREAD_GRAYSCALE)
    
    if img is None:
        raise ValueError(
            f"❌ Could not load image! Check if it's a valid image file (not corrupted, not a non-image like .txt)"
        )
    
    return img

# Load your image
try:
    original_img = load_local_image(IMAGE_PATH)
    print(f"✅ Image loaded successfully!")
    print(f"Path: {os.path.abspath(IMAGE_PATH)}")
    print(f"Resolution (Height × Width): {original_img.shape[0]} × {original_img.shape[1]}")
    print(f"Pixel value range: {original_img.min()} → {original_img.max()}")
except Exception as e:
    print(f"Error: {e}")
    # Fallback: Generate a test image if your file is missing (so code still runs)
    print("\n⚠️ Falling back to a generated test image (fix the path above to use your image!)")
    gradient = np.linspace(0, 255, 512, dtype=np.uint8)
    original_img = np.tile(gradient, (512, 1))
```

```python
# Display the original image
plt.figure(figsize=(8, 6))
plt.imshow(original_img, cmap='gray', vmin=0, vmax=255)
plt.title(f"Original Image: {os.path.basename(IMAGE_PATH)}", fontsize=12)
plt.colorbar(label="Pixel Value (0 = Black, 255 = White)")
plt.axis('off')
plt.show()
```

```python
# --------------------------
# Step 1: Define YOUR custom boundaries (edit these values!)
# --------------------------
LOWER_BOUND = 50   # Pixels below this become 50
UPPER_BOUND = 200  # Pixels above this become 200
# Optional: Round intermediate values to nearest integer (set to True/False)
ROUND_VALUES = False

# --------------------------
# Step 2: Perform quantization
# --------------------------
# Convert image to float32 to avoid uint8 overflow/underflow issues
img_float = original_img.astype(np.float32)

# Apply lower/upper bounds (core quantization logic)
quantized_img = np.clip(img_float, LOWER_BOUND, UPPER_BOUND)

# Optional: Round values (uncomment if ROUND_VALUES = True)
if ROUND_VALUES:
    quantized_img = np.round(quantized_img)

# Convert back to uint8 (standard image format)
quantized_img = quantized_img.astype(np.uint8)

# --------------------------
# Step 3: Print stats for comparison
# --------------------------
print(f"=== Quantization Results ===")
print(f"Custom Boundaries: Lower = {LOWER_BOUND}, Upper = {UPPER_BOUND}")
print(f"Original Pixel Range: {original_img.min()} → {original_img.max()}")
print(f"Quantized Pixel Range: {quantized_img.min()} → {quantized_img.max()}")

# --------------------------
# Step 4: Display original vs quantized side-by-side
# --------------------------
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Original image
ax1.imshow(original_img, cmap='gray', vmin=0, vmax=255)
ax1.set_title("Original Image", fontsize=12)
ax1.axis('off')
ax1.colorbar = plt.colorbar(ax1.imshow(original_img, cmap='gray', vmin=0, vmax=255), ax=ax1, fraction=0.046, pad=0.04)
ax1.colorbar.set_label("Pixel Value")

# Quantized image
ax2.imshow(quantized_img, cmap='gray', vmin=0, vmax=255)
ax2.set_title(f"Quantized Image (Lower={LOWER_BOUND}, Upper={UPPER_BOUND})", fontsize=12)
ax2.axis('off')
ax2.colorbar = plt.colorbar(ax2.imshow(quantized_img, cmap='gray', vmin=0, vmax=255), ax=ax2, fraction=0.046, pad=0.04)
ax2.colorbar.set_label("Pixel Value")

plt.tight_layout()
plt.show()
```
